```python
data_df = data_df.rename(columns={'batch_text' : 'sentence', 'theme':'pillar', 'batch_counter': 'sentence_number'})
data_df
```

```python
def convertSentimentToText(sentimentNumber):
    if sentimentNumber > 3 :
        sentimentText = 'Positive'
    elif sentimentNumber < 3:
        sentimentText = "Negative" 
    else :
        sentimentText = "Neutral"
    return sentimentText
```

```python
data_df['Sentiment Text'] = data_df['Sentiment Number'].apply(convertSentimentToText)
data_df
```

```python
def getOS(Brand):
    if Brand == 'iPhone':
        os = 'iPhone'
    else:
        os='Android'
    return os
```
